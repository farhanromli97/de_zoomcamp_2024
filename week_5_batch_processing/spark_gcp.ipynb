```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
from pyspark.context import SparkContext
```

```python
credential_location = "/Users/farhanromli/Library/CloudStorage/OneDrive-Personal/de_zoomcamp_2024/week_2_mage/mage-zoomcamp/dtc-de-course-412112-434d0341aa41.json"

conf = SparkConf() \
    .setMaster('local[*]') \
    .setAppName('test') \
    .set("spark.jars", "./lib/gcs-connector-hadoop3-2.2.5.jar") \
    .set("spark.hadoop.google.cloud.auth.service.account.enable", "true") \
    .set("spark.hadoop.google.cloud.auth.service.account.json.keyfile", credential_location)
```

```python
sc = SparkContext(conf=conf)

hadoop_config = sc._jsc.hadoopConfiguration()

hadoop_config.set("fs.AbstractFileSystem.gs.impl",  "com.google.cloud.hadoop.fs.gcs.GoogleHadoopFS")
hadoop_config.set("fs.gs.impl", "com.google.cloud.hadoop.fs.gcs.GoogleHadoopFileSystem")
hadoop_config.set("fs.gs.auth.service.account.json.keyfile", credential_location)
hadoop_config.set("fs.gs.auth.service.account.enable", "true")
```

```python
spark = SparkSession.builder \
    .config(conf=sc.getConf()) \
    .getOrCreate()
```

```python
df_green = spark.read.parquet('gs://mage-zoomcamp-2024-farhan-romli/pq/green/*/*') 
```
